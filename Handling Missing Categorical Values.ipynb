```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"DataSets\loan.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.fillna(10).head(10)
# It's very wrong way to fill Missing values
```

```python
# Two main types of data Numerical and Categorical/sting/object,
# There are three right ways to fill missing categorical data.
# 1. Mod
# 2. Backward Filling
# 3. Farward Filling
```

```python
# 8/4/2024
# Backward and Forward Filling
# df.fillna(method = "bfill").head(10)
# df.fillna(method = "ffill").head(10)

# Here, we can use axis to fill Row or Column wise

df["LoanAmount"].fillna(method = "bfill", axis = 0, inplace = True)
```

```python
df.isnull().sum()
```

```python
# Check Dataset Information
df.info()
```

```python
# Filling Miising Categorical Data in a Column "BY USING MOD"
mostRepeatedValue = df["Gender"].mode()[0]
```

```python
df["Gender"].fillna(mostRepeatedValue, inplace=True)
```

```python
df.isnull().sum()
```

```python
# If you want to add MOD to all Object type Column. You can use for loop for this

# df.select_dtypes(include = "object").isnull().sum() # Instead using this you can do this thing like this;

colNames = df.select_dtypes(include = "object").columns

for colName in colNames:
    mod = df[colName].mode()[0]
    df[colName].fillna(mod, inplace = True)
```
